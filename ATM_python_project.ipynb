```python
import pandas as pd
from datetime import datetime as dt
```

```python
#@title Class Function
class Account:
  def __init__(self,name,password):
    self.name = name
    self.balance = 0
    self.password = str(password)
    self.history = []
    self.time = []
    self.activate = dt.now()
  
  def check_balance(self):
    message = f"Account: {self.name} , Balance: {self.balance}"
    print(message)

  def deposit(self, money):
    #identify
    passw = str(input("Input Password : "))
    if passw != self.password:
      print("Password failed")
    else:
    #operate
      self.balance += money
      deposit = "Deposit " + str(money)
      self.history.append(deposit)

      now = dt.now().strftime("%d/%m/%Y %H:%M:%S")
      self.time.append(now)
      print(f"New balance : {self.balance}")
      print("Deposit successfully")
  
  def transfer(self, money ,name):
    #identify
    passw = str(input("Input Password : "))
    if passw != self.password:
      print("Password failed")
    else:
    #operate
      self.balance -= money
      transfer = "Transfer " + str(money) + " to " +  str(name)
      self.history.append(transfer)

      now = dt.now().strftime("%d/%m/%Y %H:%M:%S")
      self.time.append(now)
      print(f"New balance : {self.balance}")
      print(f"Transfer successfully")


  def withdraw(self, money):
    #identify
    passw = str(input("Input Password : "))
    if passw != self.password:
      print("Password failed")
    else:
    #operate
      self.balance -= money
      withdraw = "Withdraw " + str(money)
      self.history.append(withdraw)

      now = dt.now().strftime("%d/%m/%Y %H:%M:%S")
      self.time.append(now)
      print(f"New balance : {self.balance}")
      print(f"Withdraw successfully")

  def record(self):
    df = pd.DataFrame({
        'time':self.time,
        'record':self.history
    })
    print(df)
  

  def change_name(self,newname):
    self.name = newname
    print(f"New name: {self.name}")

  def change_password(self):
    #identify
    passw = str(input("Input Password : "))
    if passw != self.password:
      print("Password failed")
    else:
      newpassw = str(input("Input New Password : "))
      self.password = newpassw
      print(f"New password successfully")
  
  #inheritance
class Employee(Account):
  def __init__(self,name,password,staff_id,branch,position):
     super().__init__(name,password)
     self.position = position
     self.branch = branch
     self.id = staff_id
    
  def bonus(self):
     pass

def press_number(list):
  while True:
    choice = int(input("press number:"))
    if choice not in list:
      print("Try again!")
    else:
      break
  return choice

def registeration_normal ():
  var1 = str(input("Name:"))
  var2 = str(input("Password:"))
  name = Account(var1,var2)
  return name

def registeration_staff ():
  var1 = str(input("Name:"))
  var2 = str(input("Password:"))
  var3 = str(input("Staff_id:"))
  var4 = str(input("Branch:"))
  var5 = str(input("Position:"))
  name = Employee(var1,var2,var3,var4,var5)
  return name

def match(name,list):
  i = 0
  for x in list:
    if x.name == name:
      break
    else:
      i += 1
  return i
```
