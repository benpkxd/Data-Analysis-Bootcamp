```python
# loop for data and create data frame

first_names = []
last_names = []
heights = []
positions = []
team_full_names = []
team_cities = []
team_divisions = []

for i in range(1,101):
  url = f"https://free-nba.p.rapidapi.com/players/{i}"
  response = requests.request("GET", url, headers=headers)
  result = response.json()
  first_names.append(result["first_name"])
  last_names.append(result["last_name"])
  heights.append(result["height_feet"])
  positions.append(result["position"])
  team_full_names.append(result["team"]["full_name"])
  team_cities.append(result["team"]["city"])
  team_divisions.append(result["team"]["division"])
  time.sleep(1)

nba_players = pd.DataFrame({
    "first_name": first_names,
    "last_name": last_names,
    "height": heights,
    "position": positions,
    "team_full_name": team_full_names,
    "team_city": team_cities,
    "team_divisions": team_divisions
})

nba_players
```
