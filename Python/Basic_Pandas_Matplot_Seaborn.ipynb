```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/kaggle/input/top-300-asian-universities-qs-rankings-2024/topuniversities.csv")
```

```python
df=df.dropna()
```

```python
#EDA
df.describe()
```

```python
Top_uni = df[df['Overall Score']==df['Overall Score'].max()]
```

```python
Top_uni
```

```python
country_n =  df['Country'].value_counts()
country_n
```

```python
#Data viz - Bar chart
plt.figure(figsize=(10,6))
sns.barplot(x=country_n.index,
           y=country_n.values,
           palette = 'viridis')
plt.title('Number of University in each country')
plt.xlabel('Country')
plt.ylabel('Number of U.')
plt.xticks(rotation = 40)
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(10,6))
sns.scatterplot(x = 'Citations per Paper',
               y = 'Overall Score',
               hue = 'Country',
               data = df,
               palette = 'deep')

plt.title('Relationship between citation and Overall score')
plt.xlabel('Citation per paper')
plt.ylabel('Overall score')
plt.show()
```

```python
#Box plot
plt.figure(figsize=(12,6))
sns.boxplot(x='Country',
           y='Overall Score',
           data = df,
           palette ='pastel')
plt.title('Score Distribution in each country')
plt.xlabel('Country')
plt.ylabel('Overall Score')
plt.xticks(rotation =45)
plt.show()
```
